```python
# import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn
```

```python
# read in the dataset

pd.set_option('display.max_columns', None)
df = pd.read_csv('energydata_complete.csv')
df.head()
```

```python
# print out the column names
df.columns
```

```python
# rename columns to be more descriptive
column_names = {'T1':'Kitchen_Temp','RH_1':'Kitchen_Humidity', 'T2':'Livingroom_Temp', 'RH_2':'Livingroom_Humidity',
               'T3':'Laundry_Temp','RH_3':'Laundry_Humidity', 'T4':'Office_Temp','RH_4':'Office_Humidity',
               'T5':'Bathroom_Temp','RH_5':'Bathroom_Humidity', 'T6':'Outnorth_Temp','RH_6':'Outnorth_Humidity',
               'T7':'Ironroom_Temp','RH_7':'Ironroom_Humidity','T8':'Teenroom_Temp',  'RH_8':'Teenroom_Humidity',
               'T9':'Paroom_Temp','RH_9':'Paroom_Humidity','T_out':'Outside_Temp_Station','RH_out':'Outside_Humidity_Station' }

df = df.rename(columns = column_names)
df.head()
```

```python
# Implementing simple linear regression

from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error

x = df[['Livingroom_Temp']]
y = df['Outnorth_Temp']

# split the dataset
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size= 0.3, random_state = 42)

# fit the model

slm = LinearRegression()
slm.fit(x_train, y_train)

# test prediction
y_predict = slm.predict(x_test)

# get root mean squared error

rmse = np.sqrt(mean_squared_error(y_test, y_predict))
round(rmse, 3)
```

```python
# drop the date column
df1 = df.drop(columns = ['date','lights'])

# segment the features and target variables
features_df = df1.drop(columns = 'Appliances')
target = df1['Appliances']

# split the data into training and test set

from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(features_df, target, test_size = 0.3, random_state = 42)

# normalise dataset to a common scale using minmax scaler

from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()

normalised_train = pd.DataFrame(scaler.fit_transform(x_train))
normalised_test = pd.DataFrame(scaler.fit_transform(x_test))
```

```python
from sklearn.linear_model import LinearRegression
lm = LinearRegression()
lm.fit(normalised_train, y_train) # fits the model to the training set

train_predict = lm.predict(normalised_train)
predicted_values = lm.predict(normalised_test)
```

```python
# calculate the mae of the train and test set

from sklearn.metrics import mean_absolute_error

mae_train = mean_absolute_error(y_train, train_predict)
mae_test = mean_absolute_error(y_test, predicted_values)

print(round(mae_train, 3))
print(round(mae_test, 3))
```

```python
# calculating root mean squared error

from sklearn.metrics import mean_squared_error

rmse_test = np.sqrt(mean_squared_error(y_test, predicted_values))
rmse_train = np.sqrt(mean_squared_error(y_train, train_predict))

print(round(rmse_test, 3))
print(round(rmse_train, 3))
```

```python
# implement ridge regression

from sklearn.linear_model import Ridge
ridge_reg = Ridge()
ridge_reg.fit(normalised_train, y_train)
```

```python
ridge_predict = ridge_reg.predict(normalised_test)
```

```python
rmse_test = np.sqrt(mean_squared_error(y_test, ridge_predict))
print(round(rmse_test, 3))
```

```python
# implement lasso regression

from sklearn.linear_model import Lasso
lasso_reg = Lasso()
lasso_reg.fit(normalised_train, y_train)
```

```python
lasso_predict = lasso_reg.predict(normalised_test)
```

```python
rmse_test = np.sqrt(mean_squared_error(y_test, lasso_predict))
print(round(rmse_test, 3))
```

```python
#comparing the effects of regularisation

def get_weights_df(model, feat, col_name):
  #this function returns the weight of every feature
  weights = pd.Series(model.coef_, feat.columns).sort_values()
  weights_df = pd.DataFrame(weights).reset_index()
  weights_df.columns = ['Features', col_name]
  weights_df[col_name].round(3)
  return weights_df

linear_model_weights = get_weights_df(lm, x_train, 'Linear_Model_Weight')
ridge_weights_df = get_weights_df(ridge_reg, x_train, 'Ridge_Weight')
lasso_weights_df = get_weights_df(lasso_reg, x_train, 'Lasso_weight')

final_weights = pd.merge(linear_model_weights, ridge_weights_df, on='Features')
final_weights = pd.merge(final_weights, lasso_weights_df, on='Features')
print(final_weights)
```
